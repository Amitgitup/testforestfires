# Algerian Forest Fires Dataset

Data set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria, namely Bejaia region located in the north-east of Algeria and the Sidi bel-abbes region located in the  north-west of Algeria.

12 instances for each region.

The period from June 22 to September 2012. The dataset includes 11 attributes and 1 output attribute (class). The 244 intances has been classified into fire (138 classes) and not fire (106 classes) classes.

Attribute Information:

1. Date: (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012) Weather data observations.
2. Temp: temperature noon (temperature max) in Celsius Degrees: 22 to 42
3. RH: Relative Humidity in %: 21 to 90
4. Ws: Wind speed in km/hr : 6 to 29
5. Rain: total day in mm : 0 to 16.8 FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system : 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) from the FWI system: 1.1 to 6.8
11. Fire Weather Index(FWI): 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

# Data Cleaning

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the region

1. : "Bejaia Region Dataset"

2. : "Sidi-Bel Abbes Region Dataset"

Add new column with Region 

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values 
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## Remove the 122nd row
df=df.drop([122]).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fix spaces in Column names  
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Change the required columns as integer data type

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

## Changing other columns to Float data type

```python
df.columns
```

```python
#  df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']]=df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)    

objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```
